```python
import requests as req
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

I start by defining a function to get the race results for each year and race type (half/full) from the Rock and Roll Marathon results website.  Then, putting those results into a DataFrame and combining that DataFrame with the DataFrame of all previous results for that same year/type.  Finally, check if the DataFrame is empty to see if we've gotten to the end of rsults.

```python
def getRaceResults(full, year):
    page = 1
    notEmpty = True
    fullYearDF = pd.DataFrame()
    if full:
        raceType = "Marathon"
    else:
        raceType = "Half-Marathon"
    URL = "https://www.runrocknroll.com/Events/Nashville/The-Races/"+ raceType + "/" + str(year) + "-Results"
    while notEmpty:        
        response = req.post(URL, {'page': str(page)})
        soup = BeautifulSoup(response.content, "html.parser")
        race_search_main = soup.find(attrs={"class": "race-search-main"}).find("table")
        dataframes = pd.read_html(str(race_search_main))
        df = dataframes[0]
        frames = [fullYearDF, df]
        fullYearDF = pd.concat(frames)
        page = page + 1
        if df.empty:
            notEmpty = False
    return fullYearDF
```

```python
resultsFull2016 = getRaceResults(True,2016)
resultsFull2017 = getRaceResults(True,2017)
resultsFull2018 = getRaceResults(True,2018)
resultsFull2019 = getRaceResults(True,2019)
resultsHalf2016 = getRaceResults(False,2016)
resultsHalf2017 = getRaceResults(False,2017)
resultsHalf2018 = getRaceResults(False,2018)
resultsHalf2019 = getRaceResults(False,2019)
```

I convert all race times into timedeltas instead of the more standard datetimes, since there's no date associated with race times.

```python
resultsFull2016['Time'] = pd.to_timedelta(resultsFull2016.Time)
resultsFull2017['Time'] = pd.to_timedelta(resultsFull2017.Time)
resultsFull2018['Time'] = pd.to_timedelta(resultsFull2018.Time)
resultsFull2019['Time'] = pd.to_timedelta(resultsFull2019.Time)
resultsHalf2016['Time'] = pd.to_timedelta(resultsHalf2016.Time)
resultsHalf2017['Time'] = pd.to_timedelta(resultsHalf2017.Time)
resultsHalf2018['Time'] = pd.to_timedelta(resultsHalf2018.Time)
resultsHalf2019['Time'] = pd.to_timedelta(resultsHalf2019.Time)
```

Next, I wanted to compare the amount of no-show/dropouts/DNF per year (indicated by a time of '00:00:00'), so I make new dataframes of just that data, and then compare with the size of the original dataframe to get a percentage of records that were dropout/no-show.  As we can see from the graphs, that data wasn't kept for the 2018 year, for whatever reason.

```python
dropoutsFull2016 = resultsFull2016[resultsFull2016.Time == pd.Timedelta(minutes=0)]
dropoutsFull2017 = resultsFull2017[resultsFull2017.Time == pd.Timedelta(minutes=0)]
dropoutsFull2018 = resultsFull2018[resultsFull2018.Time == pd.Timedelta(minutes=0)]
dropoutsFull2019 = resultsFull2019[resultsFull2019.Time == pd.Timedelta(minutes=0)]
dropoutsHalf2016 = resultsHalf2016[resultsHalf2016.Time == pd.Timedelta(minutes=0)]
dropoutsHalf2017 = resultsHalf2017[resultsHalf2017.Time == pd.Timedelta(minutes=0)]
dropoutsHalf2018 = resultsHalf2018[resultsHalf2018.Time == pd.Timedelta(minutes=0)]
dropoutsHalf2019 = resultsHalf2019[resultsHalf2019.Time == pd.Timedelta(minutes=0)]

pctDropoutFull2016 = (dropoutsFull2016.size/resultsFull2016.size)*100
pctDropoutFull2017 = (dropoutsFull2017.size/resultsFull2017.size)*100
pctDropoutFull2018 = (dropoutsFull2018.size/resultsFull2018.size)*100
pctDropoutFull2019 = (dropoutsFull2019.size/resultsFull2019.size)*100
pctDropoutHalf2016 = (dropoutsHalf2016.size/resultsHalf2016.size)*100
pctDropoutHalf2017 = (dropoutsHalf2017.size/resultsHalf2017.size)*100
pctDropoutHalf2018 = (dropoutsHalf2018.size/resultsHalf2018.size)*100
pctDropoutHalf2019 = (dropoutsHalf2019.size/resultsHalf2019.size)*100

dictDropFull = {'2016':pctDropoutFull2016,'2017':pctDropoutFull2017,'2018':pctDropoutFull2018,'2019':pctDropoutFull2019}
dictDropHalf = {'2016':pctDropoutHalf2016,'2017':pctDropoutHalf2017,'2018':pctDropoutHalf2018,'2019':pctDropoutHalf2019}

plt.rcParams['figure.figsize'] = [10, 5]
plt.subplot(1,2,1)
plt.title('Half M. DNF %')
plt.bar(*zip(*dictDropHalf.items()))
plt.subplot(1,2,2)
plt.title('Full M. DNF %')
plt.bar(*zip(*dictDropFull.items()))
plt.show()
```

I then wanted to show frequency of finishing times in each year/type combination, excluding those represented in the previous graphs that did not finish.  To start, it was necessary to write a custom function to format the x-axis of each graph into a more readable form, since timedelta is in number of nanoseconds.

```python
#from matplotlib.ticker import FuncFormatter
import matplotlib.ticker as ticker
def format_func(x, pos):
    hours = int(x//3600)
    minutes = int((x%3600)//60)
    seconds = int(x%60)

    return "{:d}:{:02d}".format(hours, minutes)
```

After creating the new dataframes with only finished times, I loop through each column and row of the desired figure, creating a histogram subplot for each race type/year, and pass it the formatter function written above.

```python
finishedFull2016 = resultsFull2016[resultsFull2016.Time > pd.Timedelta(seconds=0)]
finishedFull2017 = resultsFull2017[resultsFull2017.Time > pd.Timedelta(seconds=0)]
finishedFull2018 = resultsFull2018[resultsFull2018.Time > pd.Timedelta(seconds=0)]
finishedFull2019 = resultsFull2019[resultsFull2019.Time > pd.Timedelta(seconds=0)]
finishedHalf2016 = resultsHalf2016[resultsHalf2016.Time > pd.Timedelta(seconds=0)]
finishedHalf2017 = resultsHalf2017[resultsHalf2017.Time > pd.Timedelta(seconds=0)]
finishedHalf2018 = resultsHalf2018[resultsHalf2018.Time > pd.Timedelta(seconds=0)]
finishedHalf2019 = resultsHalf2019[resultsHalf2019.Time > pd.Timedelta(seconds=0)]

finishedFullList = [finishedFull2016, finishedFull2017, finishedFull2018, finishedFull2019]
finishedHalfList = [finishedHalf2016, finishedHalf2017, finishedHalf2018, finishedHalf2019]

formatter = ticker.FuncFormatter(format_func)
plotNum = 1
plt.rcParams['figure.figsize'] = [15, 7]
for i in range(2):
    for j in range(4):
        ax = plt.subplot(2, 4, plotNum)
        ax.xaxis.set_major_formatter(formatter)
        ax.xaxis.set_major_locator(ticker.MultipleLocator(base=3600))
        if i == 1:
            ax.set_ylim([0,1100])
            ax.hist(finishedFullList[j].Time.astype('timedelta64[s]'),range=(0,25200))
            ax.set_title(str(2016+j) + ' Full M. Time Freq.')
        else:
            ax.set_ylim([0,9500])
            ax.hist(finishedHalfList[j].Time.astype('timedelta64[s]'),range=(0,25200))
            ax.set_title(str(2016+j) + ' Half M. Time Freq.')
        plotNum += 1
plt.tight_layout()
plt.show()
```

Finally, I want to plot the top 5 times of each race year/type, so I use a similar loop structure as in the previous code to create a figure with 8 graphs.  First though, I need to create new dataframes with just the top 5 from each.  From the results, we can see that Travis Peruski is a huge outlier in terms of finish time, and in fact his recorded time is well below the world record time for a full marathon.

```python
top5Full2016 = resultsFull2016.head(5)
top5Full2017 = resultsFull2017.head(5)
top5Full2018 = resultsFull2018.head(5)
top5Full2019 = resultsFull2019.head(5)
top5Half2016 = resultsHalf2016.head(5)
top5Half2017 = resultsHalf2017.head(5)
top5Half2018 = resultsHalf2018.head(5)
top5Half2019 = resultsHalf2019.head(5)

top5FullList = [top5Full2016,top5Full2017,top5Full2018,top5Full2019]
top5HalfList = [top5Half2016,top5Half2017,top5Half2018,top5Half2019]

formatter = ticker.FuncFormatter(format_func)
plotNum = 1
plt.rcParams['figure.figsize'] = [15, 7]
for i in range(2):
    for j in range(4):
        ax = plt.subplot(2, 4, plotNum)
        ax.xaxis.set_major_formatter(formatter)
        ax.set_yticks(np.arange(5))
        if i == 1:
            people = top5FullList[j].Name
            ax.barh(np.arange(5),top5FullList[j].Time.astype('timedelta64[s]'))            
            ax.set_title(str(2016+j) + ' Full M. Time Freq.')
        else:
            people = top5HalfList[j].Name
            ax.barh(np.arange(5),top5HalfList[j].Time.astype('timedelta64[s]'))
            ax.set_title(str(2016+j) + ' Half M. Time Freq.')
        ax.set_yticklabels(people)
        plotNum += 1
        
plt.tight_layout()
plt.show()
```
